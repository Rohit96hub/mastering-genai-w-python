```python
# A class to represent a Webpage
# If you're not familiar with Classes, check out the "Intermediate Python" notebook

# Some websites need you to use proper headers when fetching them:
headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

class Page:

    def __init__(self, url):
        """
        Create this Page object from the given url using the BeautifulSoup library
        """
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)
```

```python
page = Page("https://www.hrdive.com/news/treat-ai-as-a-new-kind-of-talent-productivity-gains/736724/")
```

```python
print(page.title)
```

```python
print(page.text)
```

```python
system_prompt = "You are an assistant that who can analyse a web page \
and provides a short summary, ignoring text that might be navigation related. \
Respond in markdown."
```

```python
user_prompt = f"You are looking at a web page titled {page.title}"
user_prompt += "\nThe contents of this website is as follows; \
please provide a short summary of this webpage in markdown. \n\n"
user_prompt += page.text
print(user_prompt)
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
print(messages)
```

```python
response = openai.chat.completions.create(
        model = "gpt-4o-mini",
        messages = messages
    )
```

```python
print(response)
```

```python
print(response.choices[0].message.content)
```
