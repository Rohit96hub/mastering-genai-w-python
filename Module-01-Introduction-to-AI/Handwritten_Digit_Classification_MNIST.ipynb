Import the 'keras' libraries and the Dataset

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
# Let's show in the second input image
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
```

```python
# Currently the pixels values are between 0 to 255. To train the Neural Network effectively we'll divide each pixel value by 255 to bring the range between 0 and 1
X_train = X_train/255
X_test = X_test/255
X_test.shape
```

```python
X_train[0]
```

```python
# Let's train the model. We'll start with keras Sequential Model
model = Sequential()
# Convert the 2D data (28 x 28) to our input for 784 by flattening it (1D)
model.add(Flatten(input_shape=(28,28)))
# Add the hidden layer with 128 Nodes and Activation function `relu`
model.add(Dense(128, activation='relu'))
# Add the output layer with softmax
model.add(Dense(10, activation='softmax'))
```

```python
# Look at the model summary
# 100480 = 784 * 128 + 128 (Biases)
model.summary()
```

```python
# Now compile the model with the given loss function
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=  ['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
y_prob
```

```python
y_pred = y_prob.argmax(axis=1)
y_pred
```

```python
# check accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Draw the loss
plt.plot(history.history['loss'])
```
