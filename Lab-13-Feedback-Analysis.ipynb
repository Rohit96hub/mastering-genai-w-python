###**Feedback Analysis**

```python
import gradio as gr
from transformers import pipeline
```

```python
# Load the Hugging Face sentiment-analysis pipeline using your specified model
pipe = pipeline("text-classification", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Define the function to process feedback
def analyze_sentiment(feedback):
    # Perform sentiment analysis
    result = pipe(feedback)[0]
    label = result['label']
    score = result['score']

    # Return the sentiment result
    # Return the feedback as Positive or Negative
    if label == "POSITIVE":
        return "<span style='color: green; font-weight: bold;'>Positive Feedback!</span>"
    else:
        return "<span style='color: red; font-weight: bold;'>Negative Feedback!</span>"
```

```python
# Create the Gradio interface
iface = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(label="Enter Feedback", placeholder="Type your feedback here..."),
    outputs="html",
    live=False
)
```

```python
# Launch the app
iface.launch()
```
